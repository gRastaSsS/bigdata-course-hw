## Report

### MVs:
* Total number of likes by user/group
* Number of comments in user's/group's posts by months

### Total number of likes by user/group

##### Main local table

Here we sort our index by liker_id to efficiently retrieve groups of likes per user.  

    CREATE TABLE IF NOT EXISTS likes_loc ON CLUSTER kube_clickhouse_cluster
    (
        item_type String,
        owner_id Int64,
        item_id Int64,
        liker_id Int64,
        ctime DateTime,
        like_date DateTime,
        post_date DateTime
    )
    ENGINE = ReplicatedMergeTree('/clickhouse/tables/{shard}/likes', '{replica}')
    PARTITION BY toYear(post_date)
    ORDER BY (liker_id);

##### Local MV

    CREATE MATERIALIZED VIEW IF NOT EXISTS likes_per_user_loc ON CLUSTER kube_clickhouse_cluster
    ENGINE = SummingMergeTree()
    ORDER BY (liker_id)
    POPULATE
    AS SELECT
      liker_id,
      count(item_id) AS number_of_likes
    FROM likes_loc
    GROUP BY liker_id;

##### Wrap local tables into Distributed. Specify hash for the main table as we will insert via Distributed.

    CREATE TABLE IF NOT EXISTS likes ON CLUSTER kube_clickhouse_cluster AS likes_loc
    ENGINE = Distributed(kube_clickhouse_cluster, default, likes_loc, xxHash64(liker_id));

    CREATE TABLE IF NOT EXISTS likes_per_user ON CLUSTER kube_clickhouse_cluster AS likes_per_user_loc
    ENGINE = Distributed(kube_clickhouse_cluster, default, likes_per_user_loc);

##### Kafka consumers

    CREATE TABLE IF NOT EXISTS likes_queue ON CLUSTER kube_clickhouse_cluster
    (
        item_type String,
        owner_id Int64,
        item_id Int64,
        liker_id Int64,
        ctime DateTime,
        like_date DateTime,
        post_date DateTime
    )
    ENGINE = Kafka('kafka-svc:9092', 'likes-topic', 'group1', 'JSONEachRow');

    CREATE MATERIALIZED VIEW IF NOT EXISTS likes_consumer TO likes AS SELECT * FROM likes_queue;

##### Final query

    SELECT liker_id, sum(number_of_likes) FROM likes_per_user GROUP BY liker_id;

### Number of comments in user's/group's posts by months

##### Main local table

We firstly sort index by date and then by from_id.

    CREATE TABLE IF NOT EXISTS posts_loc ON CLUSTER kube_clickhouse_cluster
    (
        post_id Int64,
        from_id Int64,
        comments_count Int64,
        date DateTime
    )
    ENGINE = ReplicatedMergeTree('/clickhouse/tables/{shard}/posts', '{replica}')
    ORDER BY (date, from_id);

##### Local MV

    CREATE MATERIALIZED VIEW IF NOT EXISTS comments_per_user_loc ON CLUSTER kube_clickhouse_cluster
    ENGINE = SummingMergeTree()
    ORDER BY (year, month, user_id)
    POPULATE
    AS SELECT
      from_id AS user_id,
      toYear(date) AS year,
      toMonth(date) AS month,
      sum(comments_count) AS total_comments
    FROM posts_loc
    WHERE comments_count > 0
    GROUP BY year, month, from_id;

##### Wrap local tables into Distributed. Specify hash for the main table as we will insert via Distributed.

    CREATE TABLE IF NOT EXISTS posts ON CLUSTER kube_clickhouse_cluster AS posts_loc
    ENGINE = Distributed(kube_clickhouse_cluster, default, posts_loc, xxHash64(toYear(date)));

    CREATE TABLE IF NOT EXISTS comments_per_user ON CLUSTER kube_clickhouse_cluster AS comments_per_user_loc
    ENGINE = Distributed(kube_clickhouse_cluster, default, comments_per_user_loc);
    
##### Kafka consumers

    CREATE TABLE IF NOT EXISTS posts_queue ON CLUSTER kube_clickhouse_cluster
    (
        post_id Int64,
        from_id Int64,
        comments_count Int64,
        date DateTime
    )
    ENGINE = Kafka('kafka-svc:9092', 'posts-topic', 'group-1', 'JSONEachRow');

    CREATE MATERIALIZED VIEW IF NOT EXISTS posts_consumer ON CLUSTER kube_clickhouse_cluster TO posts AS SELECT * FROM posts_queue;

##### Final query

    SELECT year, month, user_id, sum(total_comments)
    FROM comments_per_user
    GROUP BY year, month, user_id;
